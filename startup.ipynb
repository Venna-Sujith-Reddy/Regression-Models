```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("50_Startups.csv")
```

```python
# Separate features (X) and target (y)
X = df[['R&D Spend', 'Administration', 'Marketing Spend','State']]
y = df['Profit']
```

```python
X
```

```python
# One hot encode the 'State' column
X_encoded = pd.get_dummies(X, columns=['State'], drop_first=True)

# Display the encoded features
X_encoded
# Convert boolean values to 0 and 1
X_encoded = X_encoded.astype(int)
X_encoded
```

```python
import seaborn as sns
sns.pairplot(df)
```

```python
# Create correlation matrix for the 3 numerical features
correlation_matrix = df[['R&D Spend', 'Administration', 'Marketing Spend']].corr()

# Create a heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Features')
plt.show()
```

```python
# Calculate correlation between features and target
correlations = df[['R&D Spend', 'Marketing Spend', 'Profit']].corr()['Profit'].sort_values(ascending=False)
print("\nCorrelation with Profit:")
print(correlations)

# Create a bar plot to visualize correlations
plt.figure(figsize=(10, 6))
correlations[1:].plot(kind='bar')  # Exclude Profit's correlation with itself
plt.title('Feature Correlations with Profit')
plt.ylabel('Correlation Coefficient')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Import necessary libraries
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Model 1: Using all features
model1 = LinearRegression()
model1.fit(X_train, y_train)

# Make predictions
y_pred1 = model1.predict(X_test)

# Calculate metrics for model 1
r2_1 = r2_score(y_test, y_pred1)
mse_1 = mean_squared_error(y_test, y_pred1)
rmse_1 = np.sqrt(mse_1)

print("Model 1 (All Features) Results:")
print(f"R2 Score: {r2_1:.4f}")
print(f"RMSE: {rmse_1:.2f}")

# Model 2: Dropping Marketing Spend
X_no_marketing = X_encoded.drop('Marketing Spend', axis=1)
X_train2, X_test2, y_train2, y_test2 = train_test_split(X_no_marketing, y, test_size=0.2, random_state=42)

model2 = LinearRegression()
model2.fit(X_train2, y_train2)

# Make predictions
y_pred2 = model2.predict(X_test2)

# Calculate metrics for model 2
r2_2 = r2_score(y_test2, y_pred2)
mse_2 = mean_squared_error(y_test2, y_pred2)
rmse_2 = np.sqrt(mse_2)

print("\nModel 2 (Without Marketing Spend) Results:")
print(f"R2 Score: {r2_2:.4f}")
print(f"RMSE: {rmse_2:.2f}")

# Compare coefficients
print("\nModel 1 Coefficients:")
for feature, coef in zip(X_encoded.columns, model1.coef_):
    print(f"{feature}: {coef:.4f}")

print("\nModel 2 Coefficients:")
for feature, coef in zip(X_no_marketing.columns, model2.coef_):
    print(f"{feature}: {coef:.4f}")
```
